Calculating Mean

Method 1 - Manual Calculation

```python
population = [150,160,170,165,155]
sample = [150,170,165]

pop_mean = sum(population)/len(population)
sample_mean = sum(sample)/len(sample)

print("population mean is : ",pop_mean)
print("Sample mean is : ",sample_mean)
```

Method 2 - using numpy library

```python
import numpy as np

data = (85,90,78,88,95)

airthmatic_mean = np.mean(data)
print("airthmatic mean is : ",airthmatic_mean)
```

Method 3 - using Statistics

```python
import statistics as stats

var = [10,20,30,40,50]

stats_mean = stats.mean(var)

print("The ,mean is : ",stats_mean)
```

MEDIAN

Method 1 - numpy library

```python
data = [10,20,30,40,50]
median = np.median(data)
print("The median is :",median)
```

METHOD 2 - Statistics

```python
import statistics as stats

data = [10,20,30,40,50]
median = stats.median(data)
print("The median is :",median)
```

Method 3 - Pandas

```python
import pandas as pd

df = pd.DataFrame({
    'values':[10,20,30,40,50]
})
print(df)

median_pandas = df['values'].median()
print("The median is :",median_pandas)
```

  MODE

Method 1 - Statistics

```python
import statistics as stats

data = [10,20,30,40,50,50]
mode = stats.mode(data)
print("The mode is :",mode)
```

```python
import statistics as stats

data = [10,20,40,40,50,50]
mode = stats.multimode(data)
print("The mode is :",mode)
```

Scipy library

```python
from scipy import stats

data = [10,20,30,40,50]
mode = stats.mode(data)
print("The mode of the data is :",mode)
```

VARIANCE

USING NUMPY

```python
import numpy as np

data = [2,3,4,5,6,7,8]

sample_variance = np.var(data,ddof = 1)

print("The sample variance of data is :",sample_variance)

population_variance = np.var(data,ddof = 0)
print("The population variance of data is :",population_variance)
```
