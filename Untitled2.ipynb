Calculating Mean

Method 1 - Manual Calculation

```python
population = [150,160,170,165,155]
sample = [150,170,165]

pop_mean = sum(population)/len(population)
sample_mean = sum(sample)/len(sample)

print("population mean is : ",pop_mean)
print("Sample mean is : ",sample_mean)
```

VARIANCE

USING NUMPY

```python
import numpy as np

data = [2,3,4,5,6,7,8]

sample_variance = np.var(data,ddof = 1)

print("The sample variance of data is :",sample_variance)

population_variance = np.var(data,ddof = 0)
print("The population variance of data is :",population_variance)
```

USING PANDAS

```python
import pandas as pd

data = [2,3,4,5,6,7,8,9]
series = pd.Series(data)

sample_variance = series.var() ## sample variance by default
print("The sample variance is :",sample_variance)

pop_variance = series.var(ddof = 0)
print("The population variance is :",pop_variance)
```

USING Scipy

```python
from scipy import stats

data = [4,5,6,7,8,8,9]

sample_variance = stats.tvar(data)
print("The sample variance is :",sample_variance)

population_variance = stats.tvar(data,ddof = 0)
print("The population variance is :",population_variance)
```

STANDARD DEVIATION

USING numpy

```python
import numpy as np

data = [2,3,4,5,6,7,8,9]

sample_std = np.std(data,ddof = 1) ## ddof = 1 for sample variance
print("The sample standard deviation the data is :",sample_std)

population_std = np.std(data,ddof = 0) ## ddof = 0 for population variance
print("The population standard deviation the data is :",population_std)
```

USING pandas

```python
import pandas as pd
data = [2,3,4,5,6,7,8,9]
series = pd.Series(data)

sample_std = series.std() ## ddof = 1 for sample variance
print("The sample standard deviation is :",sample_std)

population_std = series.std(ddof = 0) ## ddof = 0 for population variance
print("The population standard deviation is :",population_std)
```

USING SCIPY

```python
from scipy import stats
import math  ## this is import for square root

data = [2,3,4,5,6,7,8]

sample_std = math.sqrt(stats.tvar(data))
print("The sample standard deviation is :",sample_std)

pop_std = math.sqrt(stats.tvar(data,ddof =0))
print("The population standard deviation is :",pop_std)
```

DATA Representations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

## Create a figure nad axes
fig, axs = plt.subplots(3,2,figsize = (14,12))
fig.suptitle("Histograms with KDE of various Distributions",fontsize = 16)

##Normal Distribution (Gaussian Distribution)
data_normal = np.random.normal(loc = 0,scale = 1,size = 1000) # Mean = 0, std Dev = 1
sns.histplot(data_normal,kde = True,ax=axs[0,0],color ='blue',stat = 'density',bins=30) ## kde = 0 for bar ke upper jo bana hai
axs[0,0].set_title('Normal Distribution')
axs[0,0].grid(True) ## grid means jo background me box bane hai unko bhi dikhane ke liye True

##Skewed distribution
data_right_skewed = np.random.exponential(scale=1,size=1000) ## Right skewed
data_left_skewed = np.random.exponential(scale=1,size=1000) ## left skewed

# Right skewed histogram
sns.histplot(data_right_skewed,kde = True,ax=axs[0,1],color='red',stat='density',bins = 30,label="Right Skewed")

## Left Skewed histogram
sns.histplot(data_left_skewed,kde=True,ax=axs[0,1],color='green',stat='density',bins=30,label='Left Skewed')
axs[0,1].set_title("SKEWED DISTRIBUTION")
axs[0,1].legend()
axs[0,1].grid(True)

## Uniform Distribution
data_uniform = np.random.uniform(low = 0,high=1,size = 1000) ## uniform Distribution between 0 and 1
sns.histplot(data_uniform,kde=True,ax = axs[1,0],color = 'purple',stat = 'density',bins = 30)
axs[1,0].set_title('uniform Distribution')
axs[1,0].grid(True)

## bimodel Distribution
data_bimodel = np.concatenate([np.random.normal(loc = 2,scale = 1,size=500),
                               np.random.normal(loc=2,scale = 1,size = 500)])
sns.histplot(data_bimodel,kde = True,ax= axs[1,1],color = 'orange',stat = 'density',bins = 30)
axs[1,1].set_title('biomodel Distribution')
axs[1,1].grid(True)

## Multimodel Distribution

data_multimodel = np.concatenate([
    np.random.normal(loc = -6,scale = 1,size = 500),
    np.random.normal(loc =0,scale = 1,size = 500 ),
    np.random.normal(loc = 6,scale = 1,size = 500)
])

sns.histplot(data_multimodel,kde = True,ax=axs[2,0],color = 'brown',stat = 'density',bins = 30)
axs[2,0].set_title("Multimodel Distribution")
axs[2,0].grid(True)

## Hide empty subplot

axs[2,1].axis('off')

## Adjust Layout
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()


```

T Test & P Test

```python
import numpy as np
from scipy import stats

# Step 1 : Create two random sample datasets
# suppose these are test score from two group(e.g.,group A,group B)

group_A = np.random.normal(75,10,30) # Group A with mean 75,std 10 and 30 samples
group_B = np.random.normal(80,12,30) # Group B with mean 80,std 12 and 30 samples

# Step 2 : perform an independent T  test
t_statsistics,p_value = stats.ttest_ind(group_A,group_B)

# step 3 : Display the p_values and t_statistics

print(f"T-Statistics: {t_statsistics}")
print(f"p-values:{p_value}")

# Step 4 : Interpretation based on p-value
alpha = 0.05 # significance Level
if p_value <= alpha :
  print("Reject the null hypothesis(Significance different between Groups)")
else:
  print("Fail to reject the null hypothesis(No significant difference between groups)")

```
